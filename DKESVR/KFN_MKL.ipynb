```python
import torch
import numpy as np
import torch.nn as nn
import torch.nn.functional as F
import pandas as pd

class DEK(nn.Module):
    def __init__(self,in_f,out_f):
        super(DEK,self).__init__()
        self.w1 = nn.Parameter(torch.rand(1,in_f *out_f))
        # w size:the Number of Reconstructed Features
        self.w = nn.Parameter(torch.rand(7))
        #self.w = nn.Parameter(torch.rand(6))
        self.in_F = in_f
        self.out_F = out_f
    
    def Polynomial(self,x,n):
        out1 = x
        out1 = out1 * self.w1
        out1 = out1
        out1 = (F.avg_pool1d(out1,kernel_size = self.in_F,stride = self.in_F))
        out1 = torch.pow(out1, n)
        return out1
    
    def forward(self,x): 
        x = x.repeat(1,self.out_F)
        out1 = self.Polynomial(x,1)
        out2 = self.Polynomial(x,2)
        out3 = self.Polynomial(x,3)
        out4 = self.Polynomial(x,-1)
        out5 = self.Polynomial(x,-2)
        out6 = self.Polynomial(x,-3)
        out7 = self.Polynomial(x,4)
        w_sum = torch.sum(self.w)
        ww = self.w / w_sum
        out = ww[0]*out1+ww[1]*out2+ww[2]*out3+ww[3]*out4+ww[4]*out5+ww[5]*out6+ww[6]*out7
#         out = ww[0]*out1+ww[1]*out2+ww[2]*out3+ww[3]*out4+ww[4]*out5+ww[5]*out6
        return out
    
# D_out: Reconstructed Feature Dimensionality        
D_in, D_out = 149,400

class Model(nn.Module):    
    def __init__(self):
        super(Model,self).__init__()
        self.Model = DEK(D_in,D_out)
    def forward(self, x):
        x = self.Model(x)
        x = F.leaky_relu(x)
        return x
    
model = Model()

X_train = np.loadtxt('X_train.txt')
Y_train = np.loadtxt('Y_train.txt')
X_test = np.loadtxt('X_test.txt')
Y_test = np.loadtxt('Y_test.txt')
X_train = torch.tensor(X_train, dtype=torch.float32)
y_train = torch.tensor(Y_train, dtype=torch.float32)
X_test = torch.tensor(X_test, dtype=torch.float32)
y_test = torch.tensor(Y_test, dtype=torch.float32)
Kernel_train=model(X_train)
Kernel_train_np=Kernel_train.detach().numpy()
Kernel_test=model(X_test)
Kernel_test_np=Kernel_test.detach().numpy()
np.savetxt('Kernel_train.txt', Kernel_train_np, fmt='%f')
np.savetxt('Kernel_test.txt', Kernel_test_np, fmt='%f')
torch.save(model, 'DEKN.pt')
```

```python
import torch
import numpy as np
import torch.nn as nn
import torch.nn.functional as F
import pandas as pd

class DEK(nn.Module):
    def __init__(self,in_f,out_f):
        super(DEK,self).__init__()
        self.w1 = nn.Parameter(torch.rand(1,in_f *out_f)
        self.w2 = nn.Parameter(torch.exp(-self.w1))
        self.w = nn.Parameter(torch.rand(6))
        self.ww = nn.Parameter(torch.rand(6, 6))
        self.www = nn.Parameter(torch.rand(6))
        self.in_F = in_f
        self.out_F = out_f
    
    def Gaussian(self,x,n):
        out3 = torch.exp(x)
        out3 = out3 * self.w2
        out3 = torch.pow(torch.log(out3), 2)
        out3 = (F.avg_pool1d(out3,kernel_size = self.in_F,stride = self.in_F))
        out3=-n*out3
        out3 = torch.exp(out3)
        return out3
    
    def forward(self,x): 
        x = x.repeat(1,self.out_F)
        out1 = self.Gaussian(x,1)
        out2 = self.Gaussian(x,2)
        out3 = self.Gaussian(x,3)
        out4 = self.Gaussian(x,4)
        out5 = self.Gaussian(x,5)
        out6 = self.Gaussian(x,6)
        ww1 = self.ww / torch.sum(self.ww, dim=1, keepdim=True)
        outt1 = ww1[0][0]*out1+ww1[0][1]*out2+ww1[0][2]*out3+ww1[0][3]*out4+ww1[0][4]*out5+ww1[0][5]*out6
        outt2 = ww1[1][0]*out1+ww1[1][1]*out2+ww1[1][2]*out3+ww1[1][3]*out4+ww1[1][4]*out5+ww1[1][5]*out6
        outt3 = ww1[2][0]*out1+ww1[2][1]*out2+ww1[2][2]*out3+ww1[2][3]*out4+ww1[2][4]*out5+ww1[2][5]*out6
        outt4 = ww1[3][0]*out1+ww1[3][1]*out2+ww1[3][2]*out3+ww1[3][3]*out4+ww1[3][4]*out5+ww1[3][5]*out6
        outt5 = ww1[4][0]*out1+ww1[4][1]*out2+ww1[4][2]*out3+ww1[4][3]*out4+ww1[4][4]*out5+ww1[4][5]*out6
        outt6 = ww1[5][0]*out1+ww1[5][1]*out2+ww1[5][2]*out3+ww1[5][3]*out4+ww1[5][4]*out5+ww1[5][5]*out6
        www1 = self.www/torch.sum(self.www)
        out = www1[0]*outt1+www1[1]*outt2+www1[2]*outt3+www1[3]*outt4+www1[4]*outt5+www1[5]*outt6
        return out
    
    
D_in, D_out = 149,400

class Model(nn.Module):    
    def __init__(self):
        super(Model,self).__init__()
        self.Model = DEK(D_in,D_out)
    def forward(self, x):
        x = self.Model(x)
        x = F.leaky_relu(x)
        return x
    
model = Model()

X_train = np.loadtxt('X_train.txt')
Y_train = np.loadtxt('Y_train.txt')
X_test = np.loadtxt('X_test.txt')
Y_test = np.loadtxt('Y_test.txt')
X_train = torch.tensor(X_train, dtype=torch.float32)
y_train = torch.tensor(Y_train, dtype=torch.float32)
X_test = torch.tensor(X_test, dtype=torch.float32)
y_test = torch.tensor(Y_test, dtype=torch.float32)
Kernel_train=model(X_train)
Kernel_train_np=Kernel_train.detach().numpy()
Kernel_test=model(X_test)
Kernel_test_np=Kernel_test.detach().numpy()
np.savetxt('Kernel_train.txt', Kernel_train_np, fmt='%f')
np.savetxt('Kernel_test.txt', Kernel_test_np, fmt='%f')
torch.save(model, 'DEKN.pt')
```
